**Chapter-2 Hipotes Testing**

1. T-testi
Amaç:
İki grup arasında bir değişkenin ortalamalarının anlamlı bir şekilde farklı olup olmadığını test eder.

Hedef:
Null Hipotez (H₀): Özgürlük (freedom_to_make_life_choices) ile iki bölge arasındaki ortalama fark anlamlı değildir.
Alternatif Hipotez (H₁): Özgürlük ile iki bölge arasındaki ortalama fark anlamlıdır.

```python
# T-Testi
from scipy.stats import ttest_ind
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Faktörleri belirleme
factors = [
    "logged_gdp_per_capita",
    "social_support",
    "healthy_life_expectancy",
    "freedom_to_make_life_choices",
    "generosity",
    "perceptions_of_corruption"
]

# Tüm faktörler için sonuçları saklamak için liste
results = []

# Her faktör için T-Testi
for factor in factors:
    # İki bölge seçimi
    region1 = df[df['regional_indicator'] == 'Western Europe'][factor]
    region2 = df[df['regional_indicator'] == 'North America and ANZ'][factor]

    # Eksik değerleri kaldırma
    region1 = region1.dropna()
    region2 = region2.dropna()

    # T-Testi
    t_stat, p_val = ttest_ind(region1, region2, equal_var=False)

    # Sonuçları saklama
    results.append({
        "Factor": factor,
        "T-statistic": t_stat,
        "P-value": p_val,
        "Conclusion": "H₁: Significant" if p_val < 0.05 else "H₀: Not Significant"
    })

# Sonuçları bir tabloya dönüştür
results_df = pd.DataFrame(results)

# Tabloyu görüntüleme
print("\nHypothesis Test Results for All Factors:")
results_df
```

```python
plt.figure(figsize=(20, 15))

# Her faktör için KDE plotları oluştur
for i, factor in enumerate(factors, start=1):
    plt.subplot((len(factors) + 2) // 3, 3, i)  # 3 sütunlu düzenleme
    region1_data = df[df['regional_indicator'] == 'Western Europe'][factor].dropna()
    region2_data = df[df['regional_indicator'] == 'North America and ANZ'][factor].dropna()

    sns.kdeplot(region1_data, label="Western Europe", shade=True, color="blue")
    sns.kdeplot(region2_data, label="North America and ANZ", shade=True, color="red")

    plt.title(f"{factor.replace('_', ' ').title()} Comparison")
    plt.xlabel(factor.replace('_', ' ').title())
    plt.ylabel("Density")
    plt.legend(title="Region")

plt.tight_layout()
plt.show()

```

```python
import pandas as pd
from scipy.stats import ttest_ind
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load datasets
norway_past = pd.read_csv("data/whr_200522.csv")
norway_current = pd.read_csv("data/2023.csv")

# Filter Norway's data
norway_past = norway_past[norway_past['Country name'] == "Norway"]
norway_current = norway_current[norway_current['Country name'] == "Norway"]

# Rename columns for consistency
norway_current.rename(columns={
    "Logged GDP per capita": "Log GDP per capita",
    "Healthy life expectancy": "Healthy life expectancy at birth"
}, inplace=True)

# Ensure the results list is reset
results = []

# Factors to analyze
factors = [
    "Happiness score", "Log GDP per capita", "Social support",
    "Healthy life expectancy at birth", "Freedom to make life choices",
    "Generosity", "Perceptions of corruption"
]

# Combine 2021, 2022, and 2023 data for Norway
recent_years = norway_past[norway_past['year'] >= 2021]
combined_recent = pd.concat([recent_years, norway_current])

# Perform T-tests for each factor
for factor in factors:
    # Recent (2021-2023) values
    recent_values = combined_recent[factor].replace([np.inf, -np.inf], np.nan).dropna()
    
    # Past (before 2021) values
    past_values = norway_past[norway_past['year'] < 2021][factor].replace([np.inf, -np.inf], np.nan).dropna()
    
    # Debugging: Print sizes of data
    print(f"Factor: {factor}, Recent size: {len(recent_values)}, Past size: {len(past_values)}")
    
    # Check if sufficient data exists for T-test
    if len(recent_values) < 2 or len(past_values) < 2:
        results.append({"Factor": factor, "T-statistic": None, "P-value": None, "Conclusion": "Insufficient data"})
    else:
        # Perform T-test
        t_stat, p_val = ttest_ind(past_values, recent_values, equal_var=False)
        results.append({
            "Factor": factor,
            "T-statistic": t_stat,
            "P-value": p_val,
            "Conclusion": "H₁: Significant" if p_val < 0.05 else "H₀: Not Significant"
        })

# Convert results to DataFrame
results_df = pd.DataFrame(results)

# Display results
print("\nT-Test Results for Norway (2021-2023 vs Past Years):")
print(results_df)

# Visualization
plt.figure(figsize=(12, 8))
sns.barplot(
    data=results_df,
    x="Factor",
    y="P-value",
    hue="Conclusion",
    dodge=False,
    palette="coolwarm"
)
plt.axhline(y=0.05, color='red', linestyle='--', label="Significance Threshold (p=0.05)")
plt.title("P-values of T-tests for Factors (Norway: 2021-2023 vs Past Years)", fontsize=14, fontweight='bold')
plt.xlabel("Factor", fontsize=12)
plt.ylabel("P-value", fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.legend(title="Hypothesis Test Conclusion")
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.impute import SimpleImputer

# Handle missing values by imputing with the mean
imputer = SimpleImputer(strategy='mean')
X_train_imputed = imputer.fit_transform(X_train_scaled)
X_test_imputed = imputer.transform(X_test_scaled)

# Create binary target based on the median happiness score
median_happiness = Y_train.median()
Y_train_binary = (Y_train > median_happiness).astype(int)
Y_test_binary = (Y_test > median_happiness).astype(int)

# Logistic Regression Model
logistic_model = LogisticRegression(random_state=42)
logistic_model.fit(X_train_imputed, Y_train_binary)

# Predict probabilities for the test set
Y_pred_logistic = logistic_model.predict_proba(X_test_imputed)[:, 1]

# Convert predicted probabilities to continuous values for regression evaluation
Y_pred_logistic_continuous = Y_pred_logistic * (Y_train.max() - Y_train.min()) + Y_train.min()

# Logistic Regression Metrics
mae_logistic = mean_absolute_error(Y_test, Y_pred_logistic_continuous)
mse_logistic = mean_squared_error(Y_test, Y_pred_logistic_continuous)
r2_logistic = r2_score(Y_test, Y_pred_logistic_continuous)

# Print Logistic Regression Evaluation Metrics
print("Logistic Regression Evaluation Metrics:")
print(f"Mean Absolute Error (MAE): {mae_logistic:.4f}")
print(f"Mean Squared Error (MSE): {mse_logistic:.4f}")
print(f"R-squared (R²): {r2_logistic:.4f}")

# Plot Logistic Regression Predictions
plt.figure(figsize=(10, 6))
plt.scatter(Y_test, Y_pred_logistic_continuous, alpha=0.7, color="purple", label="Predictions")
plt.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], color="red", linestyle="--", label="Perfect Prediction")
plt.title("Logistic Regression: Actual vs Predicted")
plt.xlabel("Actual Happiness Score")
plt.ylabel("Predicted Happiness Score")
plt.legend()
plt.show()
```
